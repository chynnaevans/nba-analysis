```python
# User input
player_name = 'Patty Mills'
```

```python
import pandas as pd
import json
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
from Naked.toolshed.shell import muterun_js
```

```python
response = muterun_js('importShotsData_python.js', '"{}"'format(player_name))

if response.exitcode == 0:
    file_name = response.stdout.decode('utf-8').replace('\n','')
else:
    print("call failed with err: " + response.stderr)
```

```python
def extract_season(row):
    return int(row['gameDate'][0:4])
```

```python
with open(file_name) as f:
    df = json.load(f)
    
shots = df['shot_Chart_Detail']
pd_shots = pd.DataFrame(shots)

pd_shots['year'] = pd_shots.apply(lambda row: extract_season(row), axis=1)
```

```python
pd_shots['year'].value_counts()
```

```python
fig = plt.gcf()
sns.set()

shot_scatter = sns.scatterplot(x='locX', y='locY',
                    hue='shotMadeFlag', size='year',
                    sizes=(20,70), data=pd_shots
                   )
shot_scatter.set_title("{} Total Attempts, career".format(player_name), size=25)
plt.ylim(-40, 470)
plt.xlim(-250, 250)
fig.set_size_inches(5*1.5,5.1*1.5)
```

```python
fig = plt.gcf()
sns.set()

shot_scatter = sns.scatterplot(x='locX', y='locY',
                    hue='shotMadeFlag',
                    data=pd_shots.query('year >= 2019')
                   )
shot_scatter.set_title("{} Total Attempts, 2019-20".format(player_name), size=25)
plt.ylim(-40, 470)
plt.xlim(-250, 250)
fig.set_size_inches(5*1.5,5.1*1.5)
```

```python
fig = plt.gcf()
sns.set()

shot_scatter = sns.scatterplot(x='locX', y='locY',
                    hue='shotMadeFlag',
                    data=pd_shots.query('year >= 2010 & year <= 2013')
                   )
shot_scatter.set_title("{} Total Attempts, 2010-13".format(player_name), size=25)
plt.ylim(-40, 470)
plt.xlim(-250, 250)
fig.set_size_inches(5*1.5,5.1*1.5)
```

```python
# Total attempts
# 2019-20
plt.figure(1)
fig = plt.gcf()
cmap = sns.cubehelix_palette(light=1, as_cmap=True)
attempt_map = sns.kdeplot(pd_shots.query('year >= 2019 & year <= 2020')['locX'], pd_shots.query('year >= 2019 & year <= 2020')['locY'], shade=True, cmap=cmap)

attempt_map.set_title("{} Total Attempts, 2019-20".format(player_name), size=25)

plt.ylim(-40, 470)
plt.xlim(-250, 250)
fig.set_size_inches(5*1.5,5.1*1.5)

# Career
plt.figure(2)
fig = plt.gcf()
cmap = sns.cubehelix_palette(light=1, as_cmap=True)
attempt_map = sns.kdeplot(pd_shots['locX'], pd_shots['locY'], shade=True, cmap=cmap)

attempt_map.set_title("{} Total Attempts, Career".format(player_name), size=25)

plt.ylim(-40, 470)
plt.xlim(-250, 250)
fig.set_size_inches(5*1.5,5.1*1.5)

# Shots made
# 2019-20
plt.figure(3)
fig = plt.gcf()
cmap = sns.cubehelix_palette(rot=-0.5, light=1, as_cmap=True)
made_map = sns.kdeplot(pd_shots.query('year >= 2019 & year <= 2020 & shotMadeFlag == 1')['locX'], pd_shots.query('year >= 2019 & year <= 2020 & shotMadeFlag == 1')['locY'], shade=True, cmap=cmap)

made_map.set_title("{} Shots Made, 2019-20".format(player_name), size=25)

plt.ylim(-40, 470)
plt.xlim(-250, 250)
fig.set_size_inches(5*1.5,5.1*1.5)

# Career
plt.figure(4)
fig = plt.gcf()
cmap = sns.cubehelix_palette(rot=-0.5, light=1, as_cmap=True)
made_map = sns.kdeplot(pd_shots.query('shotMadeFlag == 1')['locX'], pd_shots.query('shotMadeFlag == 1')['locY'], shade=True, cmap=cmap)

made_map.set_title("{} Shots Made, Career".format(player_name), size=25)

plt.ylim(-40, 470)
plt.xlim(-250, 250)
fig.set_size_inches(5*1.5,5.1*1.5)

# Shots missed
# 2019-20
plt.figure(5)
fig = plt.gcf()
cmap = sns.cubehelix_palette(rot=-0.1, light=1, as_cmap=True)
missed_map = sns.kdeplot(pd_shots.query('year >= 2019 & year <= 2020 & shotMadeFlag == 0')['locX'], pd_shots.query('year >= 2019 & year <= 2020 & shotMadeFlag == 0')['locY'], shade=True, cmap=cmap)

missed_map.set_title("{} Shots Missed, 2019-20".format(player_name), size=25)

plt.ylim(-40, 470)
plt.xlim(-250, 250)
fig.set_size_inches(5*1.5,5.1*1.5)

# Career
plt.figure(6)
fig = plt.gcf()
cmap = sns.cubehelix_palette(rot=-0.1, light=1, as_cmap=True)
missed_map = sns.kdeplot(pd_shots.query('shotMadeFlag == 0')['locX'], pd_shots.query('shotMadeFlag == 0')['locY'], shade=True, cmap=cmap)

missed_map.set_title("{} Shots Missed, Career".format(player_name), size=25)

plt.ylim(-40, 470)
plt.xlim(-250, 250)
fig.set_size_inches(5*1.5,5.1*1.5)
```
